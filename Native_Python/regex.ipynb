```python
import os
import re
import pandas as pd
```

```python
input_dir = r'C:\Users\Manojkumar.Rajendran\Desktop\adhoc'
```

```python
input_file = os.path.join(input_dir, 'mongo_output.txt')
```

```python
pattern_file = open(os.path.join(input_dir,input_file))
```

```python
file_contents = [line.strip() for line in pattern_file.readlines()]
```

```python
file_contents
```

```python
pattern1 = r'.*_id"\s*:\s*(\d*),.*\\t.*\\t.*\\t([a-zA-Z0-9\.\s-]*),.*'
pattern2 = r'.*_id"\s*:\s*(\d*),.*\\t.*\\t.*\\t(.*)(?:,).*'

cases_with_services = list()

n = 1
for line in file_contents:
    match = re.search(pattern1, line)
    if match is None:
        print('First pattern failed for',n)
        match = re.search(pattern2, line)
    case_id = match.groups()[0]
    service_name = match.groups()[1]
    cases_with_services.append({'case_id':case_id,'service_name':service_name})
    n += 1
x = pd.DataFrame(cases_with_services)
x.to_csv('ids_with_service_names.csv',index=False)
```
